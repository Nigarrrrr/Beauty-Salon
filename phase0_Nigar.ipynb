```python
import csv
with open("scores.csv","w") as file:
  file.write("name,score\n")
  file.write("Alice,85\n")
  file.write("Bob,72\n")
  file.write("Charkie,90\n")
total=0
numbers=0
with open("scores.csv","r") as file:
    reader=csv.reader(file)
    next(reader)
    for row in reader:
     scores = int(row[1])
     total += scores
     numbers += 1
average=total/numbers
print("average:",average)

```

Task 13

```python
import csv
def average():
  file=open('scores.csv','r')
  reader=csv.reader(file)
  total=0
  numbers=0
  for row in reader:
    if reader.line==1:
      continue
    score=int(row)
    total=total=score
    numbers=numbers+1
  file.close()
  avg=total/numbers
  return avg
def summary():
  Average=average
  output=open('summary.txt','w')
  output.write("average :"+str(Average))
  output.close()
setup = open('scores.csv', 'w', newline='')
writer = csv.writer(setup)
writer.writerow(['name', 'score'])
writer.writerow(['Alice', '85'])
writer.writerow(['Bob', '72'])
writer.writerow(['Charlie', '90'])
setup.close()
summary()
print("Done")
```
